```python
plt.pie(population_len_crore['Population(Crore)'],labels=population_len_crore["State"],autopct="%1.0f%%")
plt.title("Population in Crore ")
plt.xticks(rotations=45,ha='right')
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.bar(data['State'],data['Population(Crore)'])
plt.xticks(rotation=45,ha='right')
plt.title("State Wis Population In Crore",size=29)
plt.xlabel('Indian State',size=15)
plt.ylabel('Population in Crore',size=15)
plt.show()
```

```python
lok_sabha_ruling_total_seats=data.groupby('Ruling Party')['Lok Sabha Seats'].sum().reset_index()
display(lok_sabha_ruling_total_seats)
```

```python
lok_sabha_ruling_total_seats=lok_sabha_ruling_total_seats.sort_values(by='Lok Sabha Seats',ascending=False)
```

```python
plt.figure(figsize=(20,15))
plt.bar(lok_sabha_ruling_total_seats['Ruling Party'],lok_sabha_ruling_total_seats['Lok Sabha Seats'],color='purple')
plt.xticks(rotation=45,ha='right')
plt.title("Lok Sabha Seats Party Wise",size=25)
plt.xlabel('Indian Ruling Partys')
plt.ylabel('Lok Sabha Seats ')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.bar(data['State'],data['Youth %'],color='maroon')
plt.xticks(rotation=45,ha='right')
plt.title('State Wise Youth % ',size=25)
plt.xlabel('Indian State',size=25)
plt.ylabel('Youth Percentage',size=25)
plt.show()
```

```python
lok_seats_ruling=data.groupby("Ruling Party")["Lok Sabha Seats"].sum().reset_index()
display(lok_seats_ruling)
print("------------------------------------------------------")
plt.figure(figsize=(20,20))
plt.pie(lok_seats_ruling['Lok Sabha Seats'],labels=lok_seats_ruling['Ruling Party'],autopct="%1.1f%%")
plt.xticks(rotation=90,ha='right')
plt.title("Ruling Party With Loksabha Seats ",size=30)
plt.show()
```
